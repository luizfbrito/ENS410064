#### 1.1 Numpy arrays

```python
import numpy as np
# Criando um vetor de dados aleat√≥rios
data = np.random.rand(1000,2)

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data))

# Print de uma amostra dos dados
print(data[0:10])
print(data[0:10].shape)
```

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[:,0])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3)
ax[0].plot(data)
ax[1].plot(data[:,0])
ax[2].plot(data[:,1])
print(type(data))
```

```python
# Criando um dado 2D
data3d = np.sin(np.random.rand(1000,1000))
print(data3d)
```

```python
# Plotando um dado 3D
fig2, ax2 = plt.subplots()
ax2.pcolor(data3d)
```

```python
# Criando vetor temporal
dtime = np.arange(np.datetime64("2022-01-01"), np.datetime64("2023-01-01"), np.timedelta64(1, "h"))
print(dtime[1:10])
print(dtime.shape)
print(type(dtime))
```

#### 1.2 Pandas

```python
import pandas as pd

df = pd.DataFrame()
print(df)
```

```python
df['datetime'] = dtime
print(df)
df
```

```python
# Criando uma coluna random
df['random'] = np.random.rand(df.shape[0])
df
```

```python
df['random'][1:10] # Acessando a coluna random
```

```python
# Indexa√ß√£o
df.iloc[0,:]
```

```python
df = df.set_index(df['datetime'])
df
```

```python
# Criando colunas com ano, m√™s e dia
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
```

```python
tt = np.arange(0,df.shape[0],1)
print(tt.shape)
df['sazonal'] = (np.sin(tt/(360*4)))
df['timeSeries'] = 10*df['sazonal']+df['random']
#df['sazonal'].plot()
df['timeSeries'].plot(color='red')
```

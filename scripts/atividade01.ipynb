```python
path = 'C:/ENS410064/dados/brutos/atividade01'
```

```python
import pandas as pd
import os
files = os.listdir(path)
print(files)
```

```python
# read dataset
data = pd.read_csv(path+'/'+'LuizBrito_Atividade01'+'.csv')
dados = data

# Print the results
print(dados)
```

```python
import matplotlib.pyplot as plt
plt.bar(dados['Evento'],dados['Pg'])

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('Precipitação total (mm)')

# Display the chart
plt.show()
```

```python
## Separation of events

# Define the width of the bars (optional, for better visualization)
bar_width = 0.35

# Create the bar chart for the first set of data
plt.bar(dados['Evento'], dados['Pg'], width=bar_width, label='Chuva externa')

# Create the bar chart for the second set of data, adjusting the x positions
plt.bar([x + bar_width for x in range(len(dados['Evento']))], dados['Tf'], width=bar_width, label='Chuva interna')

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('(mm)')
plt.title('Separação dos eventos de precipitação')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
## Create a stacked bar chart

plt.bar(dados['Evento'], dados['Pg'], label='Chuva externa', width=1)
plt.bar(dados['Evento'], dados['Tf'], label='Chuva interna', width=0.8)

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('(mm)')
plt.title('Separação dos eventos de precipitação')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
## Calculation of statistics
# import numpy
import numpy as np

# Calculate the mean
mean_Pg = np.mean(dados['Pg'])
mean_Tf = np.mean(dados['Tf'])

# Calculate the standard deviation
std_dev_Pg = np.std(dados['Pg'])
std_dev_Tf = np.std(dados['Tf'])

# Calculate the variance
variance_Pg = np.var(dados['Pg'])
variance_Tf = np.var(dados['Tf'])

# import scipy.stats
import scipy.stats

# Calculate the interquartile range (IQR)
iqr_Pg = scipy.stats.iqr(dados['Pg'])
iqr_Tf = scipy.stats.iqr(dados['Tf'])

# Print the results
print(f"Mean Pg: {mean_Pg}")
print(f"Mean Tf: {mean_Tf}")
print(f"Standard Deviation Pg: {std_dev_Pg}")
print(f"Standard Deviation Tf: {std_dev_Tf}")
print(f"Variance Pg: {variance_Pg}")
print(f"Variance Tf: {variance_Tf}")
print(f"Interquartile Range (IQR) Pg: {iqr_Pg}")
print(f"Interquartile Range (IQR) Tf: {iqr_Tf}")
```

```python
# Create a bar chart
plt.bar(dados['Evento'],dados['Pg'])

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('Precipitação total (mm)')

# Add a mean line for Pg dataset
plt.axhline(mean_Pg, color='r', linestyle='--', label=f'Mean Pg: {round(mean_Pg,2)}') # the round function rounds the average to 2 decimal places

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Create a bar chart
plt.bar(dados['Evento'],dados['Tf'])

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('Precipitação interna (mm)')

# Add a mean line for Pg dataset
plt.axhline(mean_Tf, color='r', linestyle='--', label=f'Mean Tf: {round(mean_Tf,2)}')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
## Create a stacked bar chart

plt.bar(dados['Evento'], dados['Pg'], label='Chuva externa', width=1, color='b')
plt.bar(dados['Evento'], dados['Tf'], label='Chuva interna', width=0.8, color='lightgreen')

# Add labels and title
plt.xlabel('Evento')
plt.ylabel('(mm)')
plt.title('Separação dos eventos de precipitação')

# Add a mean line for each dataset
plt.axhline(mean_Pg, color='r', linestyle='--', label=f'Mean Pg: {round(mean_Pg,2)}') # the round function rounds the average to 2 decimal places
plt.axhline(mean_Tf, color='black', linestyle='--', label=f'Mean Tf: {round(mean_Tf,2)}')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```
